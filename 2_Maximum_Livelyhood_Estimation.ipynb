```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon

# -----------------------------
# Normal Distribution (MLE)
# -----------------------------

data = np.array([5.2, 4.8, 5.5, 5.0, 4.9, 5.1, 5.3, 4.7, 5.4, 5.0])

mu_mle = np.mean(data)
sigma_mle = np.sqrt(np.mean((data - mu_mle) ** 2))

print("Normal Distribution MLE Estimates:")
print("Estimated Mean (μ):", mu_mle)
print("Estimated Standard Deviation (σ):", sigma_mle)

# Ploting Histogram + Fitted Curve
x = np.linspace(min(data), max(data), 100)
pdf = norm.pdf(x, mu_mle, sigma_mle)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(data, bins=5, density=True, alpha=0.6)
plt.plot(x, pdf)
plt.title("Normal Distribution Fit")

# -----------------------------
# Exponential Distribution (MLE)
# -----------------------------

exp_data = np.array([2.1, 1.8, 3.0, 2.5, 1.2, 2.8, 1.5, 2.3, 2.0, 1.9])

lambda_mle = 1 / np.mean(exp_data)

print("\nExponential Distribution MLE Estimate:")
print("Estimated Lambda (λ):", lambda_mle)

# Ploting Histogram + Fitted Curve
x_exp = np.linspace(0, max(exp_data), 100)
pdf_exp = lambda_mle * np.exp(-lambda_mle * x_exp)

plt.subplot(1,2,2)
plt.hist(exp_data, bins=5, density=True, alpha=0.6)
plt.plot(x_exp, pdf_exp)
plt.title("Exponential Distribution Fit")

plt.tight_layout()
plt.show()
```
