# Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Normal Distribution (mean=0, std=1)
normal_samples = np.random.normal(loc=0, scale=1, size=1000)

print("First 10 Normal Samples:")
print(normal_samples[:10])
```

# Uniform Distribution

```python
# Uniform Distribution (range -5 to 5)
uniform_samples = np.random.uniform(low=-5, high=5, size=1000)

print("\nFirst 10 Uniform Samples:")
print(uniform_samples[:10])
```

# Exponential Distribution

```python
# Exponential Distribution (lambda = 1)
exponential_samples = np.random.exponential(scale=1.0, size=1000)

print("\nFirst 10 Exponential Samples:")
print(exponential_samples[:10])
```

# Poisson Distribution

```python
# Poisson Distribution (lambda = 4)
poisson_samples = np.random.poisson(lam=4, size=1000)

print("\nFirst 10 Poisson Samples:")
print(poisson_samples[:10])
```

# Visualization of the distributions

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.hist(normal_samples, bins=30)
plt.title("Normal Distribution")

plt.subplot(2,2,2)
plt.hist(uniform_samples, bins=30)
plt.title("Uniform Distribution")

plt.subplot(2,2,3)
plt.hist(exponential_samples, bins=30)
plt.title("Exponential Distribution")

plt.subplot(2,2,4)
plt.hist(poisson_samples, bins=30)
plt.title("Poisson Distribution")

plt.tight_layout()
plt.show()
```
