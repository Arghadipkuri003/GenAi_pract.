# Step 1: Importing Libraries

```python
import tensorflow as tf
import numpy as np
```

# Step 2: Creating Dataset

```python
# Sample dataset (100 samples, 4 features)
X = np.random.randn(100, 4)

# Binary labels
y = (np.sum(X, axis=1) > 0).astype(np.float32)

# Convert to tensors
X = tf.constant(X, dtype=tf.float32)
y = tf.reshape(tf.constant(y, dtype=tf.float32), (-1, 1))
```

# Step 3: Initializing Parameters

```python
input_dim = 4
hidden_dim = 8
output_dim = 1

# Weights & Biases
W1 = tf.Variable(tf.random.normal([input_dim, hidden_dim]))
b1 = tf.Variable(tf.zeros([hidden_dim]))

W2 = tf.Variable(tf.random.normal([hidden_dim, output_dim]))
b2 = tf.Variable(tf.zeros([output_dim]))
```

# Step 4: Defining Forward Pass

```python
def forward(X):
    # Hidden layer
    Z1 = tf.matmul(X, W1) + b1
    A1 = tf.nn.relu(Z1)

    # Output layer
    Z2 = tf.matmul(A1, W2) + b2
    A2 = tf.nn.sigmoid(Z2)

    return A2
```

# Step 5: Defining Loss Function

```python
def compute_loss(y_true, y_pred):
    loss = -tf.reduce_mean(
        y_true * tf.math.log(y_pred + 1e-8) +
        (1 - y_true) * tf.math.log(1 - y_pred + 1e-8)
    )
    return loss
```

# Step 6: Training Loop Using GradientTape

```python
learning_rate = 0.01
epochs = 500

for epoch in range(epochs):

    with tf.GradientTape() as tape:
        y_pred = forward(X)
        loss = compute_loss(y, y_pred)

    gradients = tape.gradient(loss, [W1, b1, W2, b2])

    # Update parameters
    W1.assign_sub(learning_rate * gradients[0])
    b1.assign_sub(learning_rate * gradients[1])
    W2.assign_sub(learning_rate * gradients[2])
    b2.assign_sub(learning_rate * gradients[3])

    if epoch % 50 == 0:
        print(f"Epoch {epoch}, Loss: {loss.numpy():.4f}")
```

# Step 7:  Model Evaluation

```python
y_pred = forward(X)
predictions = tf.cast(y_pred > 0.5, tf.float32)

accuracy = tf.reduce_mean(tf.cast(predictions == y, tf.float32))
print("Accuracy:", accuracy.numpy())
```
